## Importing Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Dataset
file_path = r"C:\Users\nehaa\Downloads\archive (3)\world_population.csv"  # Use raw string to avoid path errors
data = pd.read_csv(file_path)

# Preview the Dataset
print(data.head())  # Display the first few rows to inspect structure
```

## Bar Chart: Top 10 Countries by Population in 2022

```python
top_10_countries = data[['Country/Territory', '2022 Population']].sort_values(by='2022 Population', ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x='2022 Population', y='Country/Territory', data=top_10_countries, palette='viridis')
plt.title('Top 10 Countries by Population in 2022', fontsize=16)
plt.xlabel('Population (in billions)', fontsize=12)
plt.ylabel('Country/Territory', fontsize=12)
plt.tight_layout()
plt.show()
```

## Histogram: Distribution of Population Density

```python
# Histogram: Distribution of Population Density (per km²)

plt.figure(figsize=(10, 6))
sns.histplot(data['Density (per km²)'], bins=20, kde=True, color='blue')
plt.title('Distribution of Population Density', fontsize=16)
plt.xlabel('Density (per km²)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()
```

##  Box Plot: Growth Rate by Continent

```python
# Box Plot: Growth Rate by Continent
plt.figure(figsize=(12, 6))
sns.boxplot(x='Continent', y='Growth Rate', data=data, palette='Set2')
plt.title('Growth Rate by Continent', fontsize=16)
plt.xlabel('Continent', fontsize=12)
plt.ylabel('Growth Rate', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
